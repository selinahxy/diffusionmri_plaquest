```python
import pandas as pd
import numpy as np

# ----------------------------------------------------
# 1. Load CSV into DataFrame
# ----------------------------------------------------
df = pd.read_csv("/Volumes/Wanglab_HDD1/UTSW/Xinyue/AD_Cell_Type/Data_for_DL/MRIdata_threeB6_merged_updated_updated.csv")      # modify filename if needed

# ----------------------------------------------------
# 2. Extract first 15 columns and save individually
# ----------------------------------------------------
first15_cols = df.columns[:15]

for col in first15_cols:
    data = df[col]
    data.to_csv(f"B6_{col}.txt", index=False, header=False)
    print(f"Saved {col}.txt")

# ----------------------------------------------------
# 3. Compute Euclidean distance matrix between points
#    using columns "x_coordinate" and "y_coordinate"
# ----------------------------------------------------
coords = df[["x_coordinate", "y_coordinate"]].to_numpy()  # n × 2

# Compute n × n Euclidean distances
diff_x = coords[:, 0][:, None] - coords[:, 0][None, :]
diff_y = coords[:, 1][:, None] - coords[:, 1][None, :]
dist_matrix = np.sqrt(diff_x**2 + diff_y**2)

# ----------------------------------------------------
# 4. Save distance matrix to .txt
# ----------------------------------------------------
np.savetxt("B6_distance_matrix.txt", dist_matrix, fmt="%.6f")
print("Saved distance_matrix.txt")

```

```python
# Load table A
# -----------------------------
gene_all = pd.read_csv("/Volumes/Wanglab_HDD1/UTSW/Xinyue/AD_Cell_Type/Data_for_correlation/3B6_plaque_gene_expression_matrix_allspots.csv")
A = pd.read_csv("/Volumes/Wanglab_HDD1/UTSW/Xinyue/AD_Cell_Type/Data_for_DL/MRIdata_threeB6_merged_updated_updated.csv")

# -----------------------------
# Inner join to extract matching rows
# -----------------------------
subtable = gene_all.merge(
    A[["x_coordinate", "y_coordinate", "sampleID"]],
    on=["x_coordinate", "y_coordinate", "sampleID"],
    how="inner")

# -----------------------------
# Save extracted subtable
# -----------------------------
subtable.to_csv("3B6_plaque_gene_expression_matrix_matchedspots_updated_updated.csv", index=False)

print("Saved: subtable_matched.csv")

# -----------------------------
# Drop last three columns
# -----------------------------
subtable_trimmed = subtable.iloc[:, :-3]
# -----------------------------
# Save final table
# -----------------------------
subtable_trimmed.to_csv("3B6_plaque_gene_expression_matrix_matchedspots_Ydata_updated_updated.csv", index=False)

print("Saved: subtable_matched_trimmed.csv")
```

```python
import pandas as pd

# -----------------------------
# Load tables
# -----------------------------
df1 = pd.read_csv("/Volumes/Wanglab_HDD1/UTSW/Xinyue/AD_Cell_Type/Data_for_correlation/3B6_plaque_gene_expression_matrix_allspots.csv")
df2 = pd.read_csv("/Volumes/Wanglab_HDD1/UTSW/Xinyue/AD_Cell_Type/Data_for_DL/MRIdata_threeB6_merged_updated_updated.csv")
regions1 = pd.read_csv("/Volumes/Wanglab_HDD1/UTSW/Xinyue/AD_Cell_Type/Data_for_group_comparison/regions.csv")
regions2 = pd.read_csv("/Volumes/Wanglab_HDD1/UTSW/Xinyue/AD_Cell_Type/Data_for_plaque_comparison/regions.csv")

# -----------------------------
# Ensure integer type
# -----------------------------
df2["templateLabelABA"] = df2["templateLabelABA"].astype(int)
regions1["templateLabelABA"] = regions1["templateLabelABA"].astype(int)
regions2["templateLabelABA"] = regions2["templateLabelABA"].astype(int)


# -----------------------------
# Merge side by side (same row order)
# -----------------------------
merged = df2.merge(
    df1,
    on=["x_coordinate", "y_coordinate", "sampleID"],
    how="left")
merged.to_csv("ThreeB6_plaque_gene_MRI_merged_matchedspots_updated_updated.csv", index=False)

# -----------------------------
# Filter rows
# -----------------------------
valid_labels1 = set(regions1["templateLabelABA"])
subtable1 = merged[merged["templateLabelABA"].isin(valid_labels1)]

# -----------------------------
# Save output
# -----------------------------
subtable1.to_csv("ThreeB6_plaque_gene_MRI_merged_comparisonregions_updated_updated.csv", index=False)

print("Saved: merged_subtable_filtered.csv")
valid_labels2 = set(regions2["templateLabelABA"])
subtable2 = merged[merged["templateLabelABA"].isin(valid_labels2)]

# -----------------------------
# Save output
# -----------------------------
subtable2.to_csv("ThreeB6_plaque_gene_MRI_merged_plaquedetectionregions_updated_updated.csv", index=False)

print("Saved: merged_subtable_filtered.csv")
```
